Activity 2 – Create DataFrame

```python
import pandas as pd

data = {
    "Name": ["Amit", "Rahul", "Sneha"],
    "Age": [20, 21, 19],
    "City": ["Delhi", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)

print("DataFrame:")
print(df)
```

Activity 3 – Display First Rows

```python
print(df.head())
```

Activity 4 – Display Data Information

```python
print(df.info())
```

Activity 5 – Select a Column

```python
print(df["Name"])
```

Activity 6 – Add New Column

```python
df["Marks"] = [85, 90, 88]

print(df)
```

Activity 7 – Calculate Average Age

```python
avg_age = df["Age"].mean()

print("Average Age:", avg_age)
```

Activity 8 – Filter Data

```python
result = df[df["Age"] > 20]

print(result)
```
