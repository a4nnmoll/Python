## Loops in Python

Q1 – Factorial of a Number

```python
num = int(input("Enter a number: "))
fact = 1

for i in range(1, num + 1):
    fact = fact * i

print("Factorial:", fact)
```

Q2 – Armstrong Number

```python
num = int(input("Enter a number: "))
sum = 0
temp = num

while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10

if num == sum:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

Q3 – Fibonacci Series

```python
n = int(input("Enter number of terms: "))

a = 0
b = 1

for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```

Q4 – Prime Number Check

```python
num = int(input("Enter a number: "))

flag = False

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            flag = True
            break

if flag:
    print("Not a Prime Number")
else:
    print("Prime Number")
```

Q5 – Palindrome Number

```python
num = int(input("Enter a number: "))
temp = num
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

if temp == rev:
    print("Palindrome Number")
else:
    print("Not a Palindrome")
```

Q6 – Sum of Digits

```python
num = int(input("Enter a number: "))
sum = 0

while num > 0:
    digit = num % 10
    sum += digit
    num //= 10

print("Sum of digits:", sum)
```

Q7 – Numbers Divisible by 5 or 7 (1–100)

```python
count = 0

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1

print("\nTotal numbers:", count)
```

Q8 – Convert Lowercase to Uppercase

Q9 – Prime Numbers Between 1 and 100

```python
for num in range(2, 101):
    prime = True

    for i in range(2, num):
        if num % i == 0:
            prime = False
            break

    if prime:
        print(num, end=" ")
```

Q10 – Multiplication Table

```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "*", i, "=", num * i)
```
