## Strings and Sets in Python

Activity 1 – Count Capital Letters

```python
text = input("Enter a string: ")

count = 0
for ch in text:
    if ch.isupper():
        count += 1

print("Number of capital letters:", count)
```

Activity 2 – Count Vowels

```python
text = input("Enter a string: ")

vowels = "aeiouAEIOU"
count = 0

for ch in text:
    if ch in vowels:
        count += 1

print("Total vowels:", count)
```

Activity 3 – Print Words in Separate Lines

```python
sentence = input("Enter a sentence: ")

words = sentence.split()

for word in words:
    print(word)
```

Activity 4 – Count Occurrences of Substring

```python
string = input("Enter main string: ")
substring = input("Enter substring: ")

count = 0

for i in range(len(string) - len(substring) + 1):
    if string[i:i+len(substring)] == substring:
        count += 1

print("Occurrences:", count)
```

Activity 5 – Count Each Alphabet (Case Insensitive)

```python
text = input("Enter a string: ").lower()

count_dict = {}

for ch in text:
    if ch.isalpha():
        count_dict[ch] = count_dict.get(ch, 0) + 1

for key in sorted(count_dict):
    print(count_dict[key], key.upper())
```

Activity 6 – Count Unique Words Using Set

```python
sentence = input("Enter a sentence: ")

words = sentence.split()

unique_words = set(words)

print("Unique words:", unique_words)
print("Number of unique words:", len(unique_words))
```

Activity 7 – Fruit Sets from User

```python
n = int(input("Enter number of fruits: "))

s1 = set()
s2 = set()

print("Enter fruits for Set 1:")
for i in range(n):
    s1.add(input())

print("Enter fruits for Set 2:")
for i in range(n):
    s2.add(input())

print("Common fruits:", s1 & s2)
print("Only in Set1:", s1 - s2)
print("Total fruits:", len(s1 | s2))
```

Activity 8 – Set Operations

```python
S1 = {"Red", "yellow", "orange", "blue"}
S2 = {"violet", "blue", "purple"}

print("Union:", S1 | S2)
print("Intersection:", S1 & S2)
print("Difference S1-S2:", S1 - S2)
print("Difference S2-S1:", S2 - S1)
```
