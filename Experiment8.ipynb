## File Handling in Python

Activity 1 – Create and Write to a File

```python
file = open("sample.txt", "w")

file.write("Hello World\n")
file.write("Welcome to Python File Handling\n")

file.close()

print("File created and data written successfully.")
```

Activity 2 – Read a File

```python
file = open("sample.txt", "r")

content = file.read()

print("File Content:")
print(content)

file.close()
```

Activity 3 – Append Data to a File

```python
file = open("sample.txt", "a")

file.write("This line is added later.\n")

file.close()

print("Data appended successfully.")
```

Activity 4 – Read File Line by Line

```python
file = open("sample.txt", "r")

for line in file:
    print(line.strip())

file.close()
```

Activity 5 – Count Words in File

```python
file = open("sample.txt", "r")

text = file.read()

words = text.split()

print("Number of words:", len(words))

file.close()
```

Activity 6 – Count Characters in File

```python
file = open("sample.txt", "r")

text = file.read()

print("Number of characters:", len(text))

file.close()
```

Activity 7 – Count Lines in File

```python
file = open("sample.txt", "r")

lines = file.readlines()

print("Number of lines:", len(lines))

file.close()
```
