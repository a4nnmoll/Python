## Question 1
Write a Python function to find the maximum and minimum numbers from a sequence of numbers (without using built-in functions).

```python
def find_max_min(numbers):
    max_num = numbers[0]
    min_num = numbers[0]

    for num in numbers:
        if num > max_num:
            max_num = num
        if num < min_num:
            min_num = num

    print("Maximum:", max_num)
    print("Minimum:", min_num)

numbers = [10, 4, 7, 2, 15]
find_max_min(numbers)
```

## Question 2
Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.

```python
def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total += i**3
    return total

num = int(input("Enter a number: "))
print("Sum of cubes:", sum_of_cubes(num))
```

## Question 3
Write a Python function to print numbers from 1 to n using recursion.

```python
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n-1)
    print(n)

num = int(input("Enter n: "))
print_numbers(num)
```

## Question 4
Write a recursive function to print Fibonacci series up to n terms.

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

terms = int(input("Enter number of terms: "))

for i in range(terms):
    print(fibonacci(i), end=" ")
```

## Question 5
Write a lambda function to find the volume of a cone.

```python
import math

volume = lambda r, h: (1/3) * math.pi * r * r * h

r = float(input("Enter radius: "))
h = float(input("Enter height: "))

print("Volume of Cone:", volume(r, h))
```

## Question 6
Write a lambda function which gives tuple of maximum and minimum values from a list.

```python
numbers = [10, 6, 8, 90, 12, 56]

result = lambda x: (max(x), min(x))

print(result(numbers))
```

## Question 7
Write functions to explain:

1. Keyword Argument  
2. Default Argument  
3. Variable Length Argument

```python
# Keyword Argument
def student(name, age):
    print("Name:", name)
    print("Age:", age)

student(age=20, name="Anmol")


# Default Argument
def greet(name="Student"):
    print("Hello", name)

greet()
greet("Anmol")


# Variable Length Argument
def add_numbers(*numbers):
    total = 0
    for n in numbers:
        total += n
    print("Sum:", total)

add_numbers(2,3)
add_numbers(1,2,3,4,5)
```
