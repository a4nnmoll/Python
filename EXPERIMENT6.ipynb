## Lists, Tuples and Dictionary

Activity 1 – Count Occurrences of Values (0–3)

```python
n = int(input("Enter number of values: "))

values = []

for i in range(n):
    x = int(input("Enter value (0-3): "))
    values.append(x)

for i in range(4):
    print("Count of", i, "=", values.count(i))
```

Activity 2 – Tuple Average

```python
n = int(input("Enter number of values: "))

numbers = []

for i in range(n):
    num = float(input("Enter number: "))
    numbers.append(num)

t = tuple(numbers)

avg = sum(t) / len(t)

print("Tuple:", t)
print("Average:", avg)
```

Activity 3 – Runner-Up Score

```python
n = int(input("Enter number of students: "))

scores = list(map(int, input("Enter scores: ").split()))

unique_scores = list(set(scores))   # remove duplicates
unique_scores.sort()

if len(unique_scores) < 2:
    print("Runner-up score not possible")
else:
    print("Runner-up score:", unique_scores[-2])
```

Activity 4 – Dictionary of Persons and Cities

```python
n = int(input("Enter number of persons: "))

data = {}

for i in range(n):
    name = input("Enter name: ")
    city = input("Enter city: ")
    data[name] = city

print("Names:", data.keys())
print("Cities:", data.values())

print("Students and Cities:")
for name, city in data.items():
    print(name, "->", city)

city_count = {}

for city in data.values():
    city_count[city] = city_count.get(city, 0) + 1

print("Number of students in each city:", city_count)
```

Activity 5 – Movie Dictionary

```python
n = int(input("Enter number of movies: "))

movies = []

for i in range(n):
    print("\nEnter details for movie", i+1)

    name = input("Movie name: ")
    year = int(input("Year: "))
    director = input("Director: ")
    cost = float(input("Production cost: "))
    collection = float(input("Collection: "))

    movie = {
        "name": name,
        "year": year,
        "director": director,
        "cost": cost,
        "collection": collection
    }

    movies.append(movie)

print("\nAll Movies:")
for m in movies:
    print(m)
```
