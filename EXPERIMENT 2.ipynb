Aim

1-Arithmetic Operations

```python
x = 9
y = 7

print("Addition:", x + y)
print("Multiplication:", x * y)
print("Division:", x / y)
print("Subtraction:", x - y)
```

2-Area of Circle

```python
import math

r = float(input("Enter radius: "))
area = math.pi * r * r

print("Area of Circle:", area)
```

3-Solve (x + y)(x + y)

```python
x = 4
y = 3

result = (x + y) * (x + y)
print("Result:", result)
```

4-Hypotenuse (Pythagoras Theorem)

```python
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))

c = (a*a + b*b) ** 0.5
print("Hypotenuse:", c)
```

5-SIMPLE INTEREST

Formula:
SI = (P × R × T) / 100

```python
p = float(input("Enter Principal: "))
r = float(input("Enter Rate: "))
t = float(input("Enter Time: "))

si = (p * r * t) / 100
print("Simple Interest:", si)
```

6-Area of Triangle (Heron's Formula)

Formula:
s = (a + b + c) / 2
Area = √[s(s-a)(s-b)(s-c)]

```python
import math

a = float(input("Enter side 1: "))
b = float(input("Enter side 2: "))
c = float(input("Enter side 3: "))

s = (a + b + c) / 2
area = math.sqrt(s * (s-a) * (s-b) * (s-c))

print("Area of Triangle:", area)
```

7-Convert Seconds into Hours, Minutes, Seconds

```python
seconds = int(input("Enter total seconds: "))

hours = seconds // 3600
minutes = (seconds % 3600) // 60
remaining_seconds = seconds % 60

print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", remaining_seconds)
```

8-Swap Two Numbers (Without Third Variable)

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

a, b = b, a

print("After Swapping:")
print("a =", a)
print("b =", b)
```

9-Sum of First n Natural Numbers

```python
n = int(input("Enter n: "))

sum_n = n * (n + 1) // 2

print("Sum of first", n, "natural numbers is:", sum_n)
```

10-Truth Table for Bitwise Operators

```python
print("A B A&B A|B A^B")

for A in [0, 1]:
    for B in [0, 1]:
        print(A, B, A&B, A|B, A^B)
```

11-Left Shift and Right Shift

```python
num = int(input("Enter number: "))
shift = int(input("Enter shift value: "))

print("Left Shift:", num << shift)
print("Right Shift:", num >> shift)
```

12-Membership Operator (Number in Sequence)

```python
sequence = (10, 20, 56, 78, 89)

num = int(input("Enter number: "))

if num in sequence:
    print("Number is in sequence")
else:
    print("Number is not in sequence")
```

13-Membership Operator (Character in String)

```python
string = input("Enter string: ")
ch = input("Enter character: ")

if ch in string:
    print("Character is present in string")
else:
    print("Character is not present in string")
```
